```python
data_ed=pd.read_csv("survey-ed.csv",header=None)
```

```python
data_ed.head()
```

```python
data_ed.shape
```

```python
rec_ed=[]
for i in range(0,681):
    rec_ed.append([str(data_ed.values[i,j]) for j in range(0,12)])
    
rules_ed=apriori(rec_ed,min_support=0.3,min_confidence=0.3,min_lift=1,min_length=3)
res_ed=list(rules_ed)
```

```python
res_ed
```

```python
LHS=[]
RHS=[]
Support=[]
Confidence=[]
Lift=[]
for item in res_ed:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    try:
        RHS.append(items[1])
        print("Rule: " + items[0] + " -> " + items[1])
        #second index of the inner list
        print("Support: " + str(item[1]))

        #third index of the list located at 0th
        #of the third index of the inner list

        print("Confidence: " + str(item[2][0][2]))
        print("Lift: " + str(item[2][0][3]))
        print("=====================================")
        #print(item[1])
        Support.append(item[1])
        Confidence.append(item[2][0][2])
        Lift.append(item[2][0][3])
        LHS.append(items[0])
    except:
        pass
```

```python
df_e={'LHS':LHS,'RHS':RHS,'Support':Support,'Confidence':Confidence,'Lift':Lift}
df_ed=pd.DataFrame(df_e)
df_ed_con=df_ed.sort_values(by='Confidence',ascending=False).reset_index(drop=True)
df_ed_li=df_ed.sort_values(by='Lift',ascending=False).reset_index(drop=True)
```
